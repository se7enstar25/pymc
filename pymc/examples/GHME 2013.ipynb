```python
import dismod_mr
import pandas as pd
data = dismod_mr.data.load('data/pancreatitis/').input_data
```

```python
countries = ['CYP', 'DNK', 'ESP', 'FIN','GBR', 'ISL']
data = data[data.area.isin(countries)]

data['age'] = (data.age_start + data.age_end)/2
groups, countrylist = pd.factorize(data.area, order=countries)
data['group'] = groups
```

```python
from pymc import * 
from pymc.distributions.timeseries import *

nk = 10
knots = np.linspace(data.age_start.min(),data.age_end.max(), nk)
nc = len(countries)


def interpolate(x0,y0, x, group):
    idx = np.searchsorted(x0, x)
    dl = np.array(x - x0[idx - 1])
    dr = np.array(x0[idx] - x)
    d=dl+dr
    wl = dr/d
    
    return wl*y0[idx-1, group] + (1-wl)*y0[idx, group]
    

with Model() as model:
    coeff_sd = T('coeff_sd', 10, 1, .5**-2)
        
        
    mean_start = Normal('mean_start',-8, sd =5 )    
    sd_start = T('sd_start', 10, 4, 5**-2)
    

    init_prior = Normal.dist(mean_start, sd=sd_start)
    logy = GaussianRandomWalk('logy', sd=coeff_sd, init=init_prior, shape = (nk, nc))
    
    y = exp(logy)
    p = interpolate(knots, y, np.array(data.age), np.array(data.group))
    sd = T('sd', 10, .0015, .005)
    
    
    vals = Normal('vals', p, sd=sd, observed = data.value)
```

```python
with model:
    s = find_MAP( vars=[sd, logy])
    
    step = NUTS(scaling = s)
    trace = sample(30, step, s)
    
    s = trace[-1]
    
    step = NUTS(scaling = s)
    trace = sample(3000, step, s)
    
```

```python
figsize(15,10)
for i, country in enumerate(countries):
    subplot(2,3,i+1)
    d = data[data.area == country]
    plot(d.age, d.value, '.')
    plot(knots, np.exp(trace[logy][:,:,i]).T, color ='r', alpha =.01);
    ylim(0, np.max(d.value))
```
