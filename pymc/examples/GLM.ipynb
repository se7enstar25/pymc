# Bayesian GLMs

```python
import matplotlib.pyplot as plt
import numpy as np
from pymc import  *
import patsy
import theano
import pandas as pd
from collections import defaultdict
from statsmodels.formula.api import glm as glm_sm
import statsmodels.api as sm
```

```python
def glm(glm_descr, data, intercept_dist=None, regressor_dist=None, init=True, init_vals=None, family=None, model=None):
    """Create a GLM nodes.
    """
    from statsmodels.formula.api import glm as glm_sm
    if intercept_dist is None:
        intercept_dist = (Normal, {'mu': 0, 'tau': 1.0E-12})
    if regressor_dist is None:
        regressor_dist = (Normal, {'mu': 0, 'tau': 1.0E-12})

    dmatrix = patsy.dmatrix(glm_descr.split('~')[1], data)
    reg_names = dmatrix.design_info.column_names
    
    if init_vals is None and init:
        init_vals = glm_sm(glm_descr, data, family=family).fit().params
    else:
        init_vals = defaultdict(lambda: None)
    
    regs = []
    regs.append(intercept_dist[0](reg_names[0], 
				  testval=init_vals['Intercept'], 
				  **intercept_dist[1]))
    for reg_name in reg_names[1:]:
        regs.append(regressor_dist[0](reg_name, 
			              testval=init_vals[reg_name], 
				      **regressor_dist[1]))

    outcome = theano.dot(np.array(dmatrix), theano.tensor.stack(*regs))
    return theano.tensor.reshape(outcome, (1, -1)), regressors

def plot_posterior_predictive(trace):
    for rand_loc in np.random.randint(0, len(trace), 30):
        rand_sample = trace[rand_loc]
        plt.plot([0, 1], [rand_sample['Intercept'], 
                          rand_sample['Intercept'] + rand_sample['x']], lw=.2, c='k')
    plt.title('Posterior predictive')
```

Simple example
==============

Lets generate some data with known slope and intercept and fit a simple linear GLM.

```python
size = 50
true_intercept = 1
true_slope = 2

x = np.linspace(0, 1, size)
y = true_intercept + x*true_slope + np.random.normal(scale=.5, size=size)
```

```python
with Model() as model:
    glm_model, glm_vars = glm('y ~ x', {'x': x, 'y': y})
    sigma = Uniform('sigma', 0, 20)
    y_data = Normal('y', mu=glm_model, sd=sigma, observed=y)
    start = find_MAP(vars=[sigma])
    step = Slice(model.vars)
    trace = sample(2000, step, start)

plt.figure(figsize=(5, 5))
plt.plot(x, y, 'x')
plot_posterior_predictive(trace)
```

Robust GLM
==========

Lets try the same model but with an outlier in the data.

```python
x_out = np.append(x, [.1, .15, .2])
y_out = np.append(y, [8, 6, 9])
with Model() as model:
    glm_model, glm_vars = glm('y ~ x', {'x': x_out, 'y': y_out})
    sigma = Uniform('sigma', 0, 20)
    y_data = Normal('y', mu=glm_model, sd=sigma, observed=y_out)
    start = find_MAP(vars=[sigma])
    step = Slice(model.vars)
    trace = sample(2000, step, start)

plt.figure(figsize=(5, 5))
plt.plot(x_out, y_out, 'x')
plot_posterior_predictive(trace)
```

Because the normal distribution does not have a lot of mass in the tails, an outlier will affect the fit strongly.

Instead, we can use the student T distribution which has heavier tails and is more robust towards outliers. As the gradient of the T distribution is very slow to compute we use Slice sampling instead of NUTS here.

```python
with Model() as model_robust:
    glm_model, glm_vars = glm('y ~ x', {'x': x, 'y': y})
    sigma = Uniform('sigma', 0, 20)
    y_data = T('y', mu=glm_model, lam=sigma**-2, nu=1, observed=y)
    start = find_MAP(vars=[sigma])
    step = Slice(model_robust.vars)
    trace = sample(2000, step, start)

plt.plot(x_out, y_out, 'x')
plot_posterior_predictive(trace)
```

# Hierarchical GLM

```python
sat_data = pd.read_csv('Guber1999data.txt')
```

```python
with Model() as model_sat:
    grp_mean = Normal('grp_mean', mu=0, sd=10)
    grp_sd = Uniform('grp_sd', 0, 100) 

    # Define distributions to use as intercept and slope
    intercept = (Normal, {'mu': sat_data.sat_t.mean(), 'sd': sat_data.sat_t.std()})
    slope = (Normal, {'mu': grp_mean, 'sd': grp_sd})
    glm_model, glm_vars = glm('sat_t ~ spend + stu_tea_rat + salary + prcnt_take', 
                              sat_data,
    	                      intercept_dist=intercept, 
                              regressor_dist=slope)
    
    sigma = Uniform('sigma', 0, 100)
    y_data = Normal('y', mu=glm_model, sd=sigma, 
                    observed=sat_data.sat_t)
    start = find_MAP(vars=[grp_mean, grp_sd, sigma])
    step = NUTS(model_sat.vars, start)
    trace_sat = sample(2000, step, start)
```

```python
plt.figsize(10, 10)
traceplot(trace_sat);
```

```python
from pandas.tools.plotting import scatter_matrix
trace_df = pd.DataFrame({name: trace.vals[0] for name, trace in trace_sat.samples.iteritems()})
scatter_matrix(trace_df, figsize=(12,12));
```
